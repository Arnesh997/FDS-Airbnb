Modification
<p>8. Removing the HTML Tags from description</p>

```python
from bs4 import BeautifulSoup
# define a function to remove HTML tags from a string
def remove_html_tags(text):
    if pd.isna(text):
        return text
    soup = BeautifulSoup(text, 'html.parser')
    return soup.get_text()

# apply the function to the "description" column
listings_df['description'] = listings_df['description'].apply(remove_html_tags)

# print the updated dataframe
listings_df['description']
```

<p>7. Seperate parking from description/name</p>

```python
parking_rows = listings_df[(listings_df['name'].str.contains('parking', case=False)) | 
                           (listings_df['amenities'].apply(lambda x: 'parking' in x))]
parking_rows
```

Creating a new column in dataframe, 'Parking', containing 0/1 based on whether the property has Parking space for the customer

```python
listings_df['Parking'] = 0
listings_df.loc[parking_rows.index, 'Parking'] = 1
```
